```python
import numpy as np
```

```python
# đọc dữ liệu từ file
ratings = np.loadtxt("ratings.txt", dtype = int)
print("ratings: \n", ratings)

# số user, số movies
num_users = ratings.shape[0]
num_movies = ratings.shape[1]

print("num_users: ", num_users)
print("num_movies: ", num_movies)
```

```python
# hàm chuẩn hóa dữ liệu
def normalize_ratings(ratings):
    num_movies = ratings.shape[1]

    ratings_mean = np.zeros(shape = (1, num_movies))
    ratings_norm = np.zeros(shape = ratings.shape)

    for i in range(num_movies):
        # lấy ra các giá trị khác 0
        idx = ratings[:, i] > 0
        # tính mean của các giá trị khác 0
        ratings_mean[0, i] = np.mean(ratings[idx, i])
        # chuẩn hóa dữ liệu
        ratings_norm[idx, i] = ratings[idx, i] - ratings_mean[0, i]

    return ratings_norm, ratings_mean
```

```python
# chuẩn hóa dữ liệu
ratings, ratings_mean = normalize_ratings(ratings)
print("ratings after normalize: \n", ratings)
```

```python

# đọc dữ liệu từ file
user_features = np.loadtxt("user_features.txt")
movie_features = np.loadtxt("movie_features.txt")

print("user_features: \n", user_features)
print("movie_features: \n", movie_features)
```

```python
# hàm tính loss function 
def calc_lost(user_features, movie_features, ratings):

    # ma trận : 1 - 0 | 1 nếu đánh giá, 0 nếu chưa đánh giá
    mask = (ratings != 0) * 1

    # số lượng đánh giá
    count = 0
    for i in range(ratings.shape[0]):
        for j in range(ratings.shape[1]):
            if ratings[i, j] != 0:
                count += 1

    # tính giá trị dự đoán
    predicted = np.dot(user_features, movie_features.T)

    # tính hàm mất mát
    lost = np.sum(np.square(mask * (predicted - ratings)))

    return lost/count
```

```python
# hàm gradient descent
def gradient_descent(user_features, movie_features, ratings, learning_rate, num_iterations):
    # ma trận did_rate có giá trị 1 tại các vị trí có ratings > 0, ngược lại là 0
    did_rate = (ratings != 0) * 1
    # print("did_rate: \n", did_rate)

    # số đánh giá
    numRated = 0
    for i in range(did_rate.shape[0]):
        for j in range(did_rate.shape[1]):
            if did_rate[i, j] == 1:
                numRated += 1

    # lặp lại quá trình gradient descent
    for i in range(num_iterations):
        # tính độ lệch giữa dữ liệu thực tế và dữ liệu dự đoán
        predicted = np.dot(user_features, movie_features.T)
        # print("predicted: \n", predicted)
        difference = (ratings - predicted) * did_rate
        # print("difference: \n", difference)

        # tính gradient  cho user_features và movie_features
        user_features_grad = -2/(numRated) * np.dot(difference, movie_features) 
        movie_features_grad = -2/(numRated) * np.dot(difference.T, user_features)



        # cập nhật user_features và movie_features
        user_features -= learning_rate * user_features_grad
        movie_features -= learning_rate * movie_features_grad

        # tính giá trị hàm mất mát
        error = calc_lost(user_features, movie_features, ratings)

        if error < 1.7:
            print("iteration: ", i, "error: ", error)

        # chỉnh lại learning_rate
        # if error < 1.6:
        #     learning_rate = 1
            
     

        # nếu error < 0.001 thì dừng quá trình gradient descent
        if error < 0.001:
            print("iteration: ", i, "error: ", error)
            break

    return user_features, movie_features
```

```python
# thiết lập các tham số cho gradient descent
learning_rate = 100
num_iterations = 1000

# thực hiện gradient descent
users_features, movie_features = gradient_descent(user_features, movie_features, ratings, learning_rate, num_iterations)
```

```python
# in ra các ma trận đặc trưng của người dùng và movies
print("user features: \n", user_features)
# print("\n")
print("movie features: \n", movie_features)

# ghi dữ liệu vào file
np.savetxt("user_features_output.txt", user_features, fmt = "%.1f")
np.savetxt("movie_features_output.txt", movie_features, fmt = "%.1f")
```

```python
# bảng đánh giá dự đoán 
predicted_ratings = np.dot(user_features, movie_features.T) + ratings_mean
print("predicted ratings: \n", predicted_ratings)

# ghi dữ liệu vào file, làm tròn 1 chữ số thập phân
np.savetxt("prediction.txt", predicted_ratings, fmt = "%.1f")
```
