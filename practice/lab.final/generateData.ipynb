```python
import numpy as np
```

```python
# # Tạo dữ liệu mẫu: số hàng = số người, số cột = số movies
# kích thước của ma trận đánh giá
num_users, num_movies = (500, 100)


# Tạo ma trận ngẫu nhiên các giá trị từ 0 đến 5
ratings = np.random.randint(6, size = (num_users, num_movies))

# hàm shape trả về bộ (m, n) với m là số hàng, n là số cột
# shape[0] là số hàng, shape[1] là số cột
ratings.shape
print("ratings: \n", ratings)

# ma trận cho người dùng mới chỉ mới đánh giá 3 bộ phim
new_user_ratings = np.zeros((1, num_movies))
new_user_ratings[0, 0] = 5
new_user_ratings[0, 1] = 3
new_user_ratings[0, 2] = 1
print("new user rating: \n",new_user_ratings)


# thêm new_user_ratings vào ratings
ratings = np.vstack([ratings, new_user_ratings])
print("ratings after append: \n", ratings)

# ghi dữ liệu vào file
np.savetxt("ratings.txt", ratings, fmt = "%d")
```

```python
import numpy as np

# giả sử features = 10 (số lượng đặc trưng)
num_features = 10
num_users, num_movies = (501, 100)

# Hàm random.randn() trả về các giá trị ngẫu nhiên trong khoảng từ 0 đến 1
# Khởi tạo ma trận đặc trưng của người dùng 
user_features = np.random.randn(num_users, num_features) 
print("user features: \n", user_features)

# Khởi tạo ma trận đặc trưng của movies
movie_features = np.random.randn(num_movies, num_features)
print("movie features: \n", movie_features)

# ghi dữ liệu vào file
np.savetxt("user_features.txt", user_features, fmt = "%.1f")
np.savetxt("movie_features.txt", movie_features, fmt = "%.1f")
```
